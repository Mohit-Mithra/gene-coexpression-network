```python
import numpy as np
```

```python
matrix = (np.random.rand(15, 30) > 0.97) * 1
matrix
```

```python
graph = []
for i in range(15):
    temp = list(np.random.choice(15, np.random.randint(15, size=1) , replace=False))
    graph.append(temp)
    
graph
```

```python
#characters = 0...29
#taxons = 0....9

conflict_graph = []

gamets = [0,0,0,0]
for u in range(len(graph)):
    temp = []
    for v in graph[u]:
        u_taxon = matrix[u]
        v_taxon = matrix[v]
        
        for char in range(matrix.shape[1]):
            
            if(u_taxon[char] == 0 and v_taxon[char] == 0):
                gamets[0] = 1
            if(u_taxon[char] == 0 and v_taxon[char] == 1):
                gamets[1] = 1
            if(u_taxon[char] == 1 and v_taxon[char] == 0):
                gamets[2] = 1
            if(u_taxon[char] == 1 and v_taxon[char] == 1):
                gamets[3] = 1
        
        print(gamets)
        if(np.sum(gamets)) == 4:
            temp.append(v)
            
    
    conflict_graph.append(temp)
    
conflict_graph
```
